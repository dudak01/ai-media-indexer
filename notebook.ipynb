## О проекте

Интеллектуальный сервис для автоматической индексации локальных медиаколлекций (видео, аудио, изображения). Использует мультимодальное машинное обучение для извлечения семантических метаданных из имён файлов и содержимого, после чего обеспечивает поиск по смыслу с помощью векторной базы FAISS.

**Используемые ML-модели:**
- **DistilBERT NER** (fine-tuned) — извлечение сущностей (название, год, качество, исполнитель) из имён файлов
- **CLIP** — мультимодальный поиск изображений по текстовому запросу
- **Sentence-Transformers (MiniLM)** — мультиязычные текстовые эмбеддинги
- **FAISS** — индекс для семантического поиска по эмбеддингам
- **EasyOCR** — распознавание текста на изображениях
- **Shazam API** — аудиоидентификация

Подробное описание архитектуры см. в README.md и в тексте ВКР.

## 2. Демонстрация ML-ядра без GUI

Прямой вызов NER-модели на тестовом имени файла.

````python
from core.ner_predictor import NERPredictor

ner = NERPredictor()

test_filename = 'The.Dark.Knight.2008.1080p.BluRay.x264.mkv'
result = ner.extract_entities(test_filename)

print(f'Имя файла: {test_filename}')
print(f'Извлечённые сущности:')
for key, value in result.items():
    print(f'  {key}: {value!r}')
````

## 3. Демонстрация семантического поиска (FAISS + CLIP)

Создание векторной базы и поиск по смысловому запросу.

````python
from core.vector_db import VectorDatabase

vdb = VectorDatabase()

# Добавляем несколько тестовых описаний
vdb.add_text('фильм про космос черную дыру и гравитацию', {'name': 'Interstellar.2014.mkv'})
vdb.add_text('бэтмен джокер криминальный триллер готэм', {'name': 'The.Dark.Knight.2008.mkv'})
vdb.add_text('хакер матрица искусственный интеллект киберпанк', {'name': 'The.Matrix.1999.mkv'})

# Поиск по смыслу
query = 'кино про искусственный интеллект'
results = vdb.search(query, k=3)

print(f'Запрос: {query!r}\n')
for i, r in enumerate(results, 1):
    print(f'{i}. {r["payload"]["name"]} (сходство: {r["score"]}%)')
````
